```swift
%include "EnableIPythonDisplay.swift"
import Python
import TensorFlow

let np = Python.import("numpy")
let plt = Python.import("matplotlib.pyplot")
IPythonDisplay.shell.enable_matplotlib("inline")

import SwiftCV
print(cvVersion())
```

```swift
var imgOrig = imread("/notebooks/Tests/SwiftCVTests/fixtures/test.png")
imgOrig = cvtColor(imgOrig, nil, ColorConversionCode.COLOR_BGR2RGB)
plt.imshow(ShapedArray<UInt8>(cvMat: imgOrig)!.makeNumpyArray())
plt.show()
```

```swift
let imgResized = resize(imgOrig, nil, Size(100, 100), 0, 0, InterpolationFlag.INTER_AREA)
plt.imshow(ShapedArray<UInt8>(cvMat: imgResized)!.makeNumpyArray())
plt.show()
```

```swift
let imgBordered = copyMakeBorder(imgOrig, nil, 40, 40, 40, 40, BorderType.BORDER_CONSTANT, RGBA(0, 0, 0, 0))
plt.imshow(ShapedArray<UInt8>(cvMat: imgBordered)!.makeNumpyArray())
plt.show()
```

```swift
let imgBlurred = GaussianBlur(imgResized, nil, Size(3, 3))
plt.imshow(ShapedArray<UInt8>(cvMat: imgBlurred)!.makeNumpyArray())
plt.show()
```

```swift
let center = Size(imgOrig.cols/2, imgOrig.rows/2)
let imgWarped = warpAffine(imgOrig, nil, getRotationMatrix2D(center, 90, 2), Size(imgOrig.size))
plt.imshow(ShapedArray<UInt8>(cvMat: imgWarped)!.makeNumpyArray())
plt.show()
```
